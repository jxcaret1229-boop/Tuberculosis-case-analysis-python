**¿QUÉ ES LA TUBERCULOSIS?**

*La tuberculosis (TB) es una enfermedad infecciosa y contagiosa causada principalmente por la bacteria Mycobacterium tuberculosis. Aunque afecta mayoritariamente a los pulmones, también puede dañar otras partes del cuerpo como los riñones, la columna vertebral y el cerebro.*

**Transmisión:** Se propaga de persona a persona a través del aire cuando alguien con la enfermedad activa en sus pulmones tose, estornuda, habla o canta, liberando bacterias que otros inhalan.

**Síntomas comunes: **Incluyen tos persistente (a veces con sangre), dolor en el pecho, fatiga, pérdida de peso, fiebre y sudores nocturnos.

**Etapas de Progresión de la Enfermedad**
La tuberculosis suele avanzar a través de tres etapas principales:
MSD Manuals
MSD Manuals

*Infección Primaria:* Ocurre cuando la bacteria entra en los pulmones. El sistema inmunitario generalmente logra rodear las bacterias en "granulomas" para contenerlas.

*Infección Latente:* La bacteria permanece "dormida" en el cuerpo. La persona no tiene síntomas, no se siente enferma y no puede contagiar a otros, pero la prueba de PPD o de sangre sale positiva.
Enfermedad Activa: Si el sistema inmunitario se debilita, las bacterias se multiplican, causan daño a los tejidos y aparecen los síntomas. En esta etapa, el paciente es contagioso.

```python
# Cómo puedo convertir columnas de filas utilizando Pandas? Existe la función melt()
# Buscas la función melt en la documentación oficial
who2 = who.melt(id_vars=["country", "iso2","iso3","year"])
who2.head(100)
```

```python
# Crear una columna que contenga el genero (masculino o femenino)
# Crear una función que contenga las reglas
def asignar_genero():
  if "m" in valor:
    return "masculino"
  else:
    return "femenino"
```

```python
who2["genero"] = who2["variable"].apply(lambda x: "masculino" if "m" in x else "femenino")
who2.head(100)
```

```python
# the method of diagnosis (rel = relapse, sn = negative pulmonary smear, sp = positive pulmonary smear, ep = extrapulmonary)
def asignar_metodo(valor):
  if "rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapulmonary"
  else:
    return "to define"
```

```python
who2["metodo"] = who2["variable"].apply(asignar_metodo)
who2.head()
```

```python
# age group (014 = 0-14 yrs of age, 1524 = 15-24, 2534 = 25-34, 3544 = 35-44 years of age, 4554 = 45-54, 5564 = 55-64, 65 = 65 years or older).
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoedad"] = who2["variable"].apply(asignar_grupoedad)
who2.head()
```

```python
#remover la columna variable
who2.drop(columns=["variable"], inplace=True)
```

```python
# reproducir la columna value a casos
who2.rename(columns={"value":"cases"}, inplace=True)
```

```python
# Traer el dato de la población total
df = pd.merge(who2, pop, on=["country","year"], how="inner")
df.head(18)
```

Tasa de incidencia (por cada 100 mil habitantes) = 100,000*casos nuevos/población

```python
# Nueva tabla para calcular la incidencia
country_year = df.groupby(["country","year"], as_index=False).agg({"cases":"sum","population":"max"})
country_year["tasa_incidencia"] = 100_100 * country_year["cases"]/country_year["population"]
country_year.head(10)
```

```python
# Nueva tabla para calcular la tasa de incidencia POR AÑO
by_year=country_year.groupby("year", as_index=False).agg(
    {"cases":"sum", "population":"sum"}
)
by_year["tasa_incidencia"]= 100_000*by_year["cases"]/by_year["population"]
by_year
```
